# How to create a RISE slideshow

Here is how you create a RISE slideshow from a Jupyter Notebook.

This assumes you are using `jupyterlab-rise`.

## Setup

First, you need to have a running instance of Jupyter Lab.

An easy way to get one running is to:

- `git clone` my repo at https://github.com/angela-tran/rise-sandbox

- start up the [Dev Container](https://code.visualstudio.com/docs/devcontainers/containers#_quick-start-open-an-existing-folder-in-a-container)

- run this command:

   ```bash
   ./bin/start.sh  # starts Jupyter Lab
   ```

_All this is doing is running the following:_
```bash
cd ../notebooks
jupyter lab
```

From the Terminal output, click the link that looks something like:

```
 http://localhost:8888/lab?token=...
```

**Tip**: Although it is a little out-of-date, I found this article helpful for getting a quick "cheat sheet" for Jupyter Notebook keyboard shortcuts (see the section called "Jupyter Notebook Shortcuts):

https://www.geeksforgeeks.org/python/getting-started-with-jupyter-notebook-python/

## Getting started

Create a new Notebook, and then create some cells.

You'll probably want them to be `Markdown` cells.

Then, under the `View` menu, open the `Property Inspector`.
